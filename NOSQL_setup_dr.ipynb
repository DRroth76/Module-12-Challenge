```python
query = {'BusinessType': 'Restaurant/Cafe/Canteen'}
fields = {'BusinessTypeID': 1, 'BusinessType': 1}
results = establishment.find(query, fields)
for result in results:
    print(result)
```

```python
filter_query = {'BusinessName': 'Penang Flavours'}
update_id = {'$set': {'BusinessTypeID': 1}}

result = establishment.update_one(filter_query, update_id)
print(result)
```

```python
query = {'LocalAuthorityName': 'Dover'}
results = establishment.find(query)
for result in results:
    pprint(result)
```

```python
filter_query = {'LocalAuthorityName': 'Dover'}
establishment.delete_many(query)
results = establishment.find(query)
for result in results:
    pprint(result)
```

```python
remaining_count = establishment.count_documents(filter_query)
if remaining_count == 0:
    print('There are no more establishments with Dover as LocalAuthorityName')
else:
    print(f'There are still {remaining_count} left.')
```
